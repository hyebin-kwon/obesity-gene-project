| 컬럼명                  | 의미 설명                                                               |
| -------------------- | ------------------------------------------------------------------- |
| **ID**               | probe ID (예: `7896738`, `7896817`) → `series matrix`의 `ID_REF`와 매칭됨 |
| **GB\_LIST**         | GenBank accession number (예: `NM_001005240`, `BC118988` 등)          |
| **SPOT\_ID**         | 위치 정보 (예: `chr1:901877-910488`)                                     |
| **seqname**          | 해당 유전자가 위치한 염색체명 (예: `chr1`, `chr14`)                               |
| **RANGE\_GB**        | GenBank 염색체 위치 정보 (예: `NC_000001.10`)                               |
| **RANGE\_STRAND**    | DNA 가닥 방향 (`+` 또는 `-`)                                              |
| **RANGE\_START**     | 유전자 시작 위치 (예: `53049`)                                              |
| **RANGE\_STOP**      | 유전자 종료 위치 (예: `54936`)                                              |
| **total\_probes**    | 해당 probe에 할당된 총 탐침 수                                                |
| **gene\_assignment** | 유전자 주석 정보. `//` 사이에 유전자 이름이 들어 있으며, 복수 가능                           |
| **mrna\_assignment** | mRNA 정보 (`ENSEMBL`, `RefSeq`, `GenBank` 등)                          |
| **category**         | 주석 타입 (예: `main`, `flmrna->unmapped` 등)                             |


```python
import re

# 예시 DataFrame: GPL11532_32230_df
# gene_assignment에서 유전자 이름 추출
def extract_gene_symbol(text):
    if pd.isna(text):
        return None
    match = re.search(r'//\s*(.*?)\s*//', text)
    return match.group(1) if match else None

# 새 열 추가
platform_df= GPL11532_32230_df.copy()
platform_df['gene_symbol'] = GPL11532_32230_df['gene_assignment'].apply(extract_gene_symbol)
```

```python
import GEOparse

# GEO accession ID가 "GSE77962"인 데이터셋을 다운로드 및 파싱하여 gse 객체 생성
# 다운로드된 파일은 ./ 디렉토리에 저장됨
gse = GEOparse.get_GEO("GSE77962", destdir="./")

# 샘플 메타데이터를 저장할 빈 리스트 생성
metadata = []

# gse.gsms는 각 GSM(샘플)에 대한 정보를 담고 있는 dictionary
for gsm_id, gsm in gse.gsms.items():
    # 각 샘플의 GSM ID를 포함한 딕셔너리 생성
    sample_info = {"gsm_id": gsm_id}

    # 샘플의 특성 정보("characteristics_ch1")를 하나씩 순회
    for line in gsm.metadata.get("characteristics_ch1", []):
        # 특성 정보는 보통 "key: value" 형태로 되어 있음
        if ":" in line:
            # ":" 기준으로 key와 value 분리 (최초의 ':'만 기준)
            key, value = line.split(":", 1)

            # key 전처리: 공백 -> 언더스코어, 소문자로 통일, 괄호 제거 등
            key = key.strip().lower().replace(" ", "_").replace("(", "").replace(")", "")

            # value는 앞뒤 공백 제거
            value = value.strip()

            # 전처리한 key-value 쌍을 sample_info에 추가
            sample_info[key] = value

    metadata.append(sample_info)

metadata_df = pd.DataFrame(metadata)
```

### * non-obesity 유전자 발굴 시 정확성

SHAP은 실제로 **각 유전자 발현량과 감량량/저항성**의 상관성을 통해 영향을 분석
따라서 비만 관련 여부는 SHAP은 모름. 이건 나중에 해석할 때 연구자가 확인하는 것. SHAP 상위 유전자 중 비만유전자 아닌 것 추출	“잠재적 신규 비만 예측 유전자” 후보→ 논문 조사, KEGG pathway 매핑	기능 해석을 통한 생물학적 의미 부여|
